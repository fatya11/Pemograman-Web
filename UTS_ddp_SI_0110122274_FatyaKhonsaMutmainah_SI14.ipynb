```python
p1 = {'nama':'Budi',
    'produk':'TV',
    'jumlah': 3}
p2 = {'nama':'Ani',
    'produk':'AC',
    'jumlah': 4}
p3 = {'nama':'Siti',
    'produk':'Kulkas',
    'jumlah': 2}
p4 = {'nama':'Dewi',
    'produk':'AC',
    'jumlah': 5}
p5 = {'nama':'Andi',
    'produk':'Kulkas',
    'jumlah': 7}
p6 = {'nama':'Dedi',
    'produk':'AC',
    'jumlah': 1}
p7 = {'nama':'Sri',
    'produk':'TV',
    'jumlah': 4}
ar_pelanggan = [p1, p2, p3, p4, p5, p6, p7]

for pelanggan in ar_pelanggan:
  for key,data in pelanggan.items():
    
    print(key,data)
    
    hargaSatuan = pelanggan['produk']
    if hargaSatuan == 'TV':
      harga = 5000000
    elif hargaSatuan == 'AC':
      harga = 6000000
    else:
      harga = 7000000

  harga_kotor = pelanggan['jumlah'] * harga
  diskon = (0.05 * harga_kotor, 0.2 * harga_kotor)[pelanggan['produk'] == 'kulkas' and pelanggan['jumlah'] >= 3]
  ppn = 0.11 * (harga_kotor - diskon)
  harga_bayar = (harga_kotor + ppn) - diskon
  print('Harga Kotor\t\t:',harga_kotor,
          '\nDiskon\t\t\t:',diskon,
          '\nPPN\t\t\t:',ppn,
          '\nHarga Total\t\t:',harga_bayar,
          '\n---------------------------------')
```
